```python
def get_company(s):
    if type(s)!=float:
        f=requests.get('https://www.imdb.com/title/'+s+'/companycredits?ref_=tt_dt_co')
        f_text=f.text
        company=re.findall('<li>\n<a href=\"\/company\/.*?\"\n>(.*?)</a>',f_text)
        return '  '.join(company)
    else:
        return ''
df=pd.read_csv('doubanMovies.csv',encoding='gbk')
df['company']=None
for i in df.index:
    df['company'][i]=get_company(df['imdb'][i])
print(df)
```

```python
import pandas as pd
df=pd.read_csv('doubanMovies.csv',encoding='gbk')
```

```python
df['company']
```

```python
df.fillna('无',inplace=True)
def to_list(s):
    return list(set(s.split('  ')))
comp=df['company'].apply(to_list).tolist()
```

```python
comp=sum(comp,[])
```

```python
from collections import Counter
dict1=Counter(comp)
a=sorted(dict1.items(),key=lambda x:x[1],reverse=True)
```

```python
df1=pd.DataFrame(a,columns=['company','freq'])
```

```python
df1.to_csv('company.csv',index=False)
```
